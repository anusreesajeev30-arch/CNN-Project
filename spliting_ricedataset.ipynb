```python
import os
import shutil
import random

# CHANGE THIS PATH
source = r"C:\Users\ACER\Downloads\Rice_Image_Dataset"

# OUTPUT FOLDERS
train_dir = os.path.join(source, "train")
test_dir = os.path.join(source, "test")

split = 0.8  # 80% train, 20% test

# Create main folders
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Loop through each class
for class_name in os.listdir(source):
    class_path = os.path.join(source, class_name)

    if os.path.isdir(class_path) and class_name not in ["train", "test"]:
        images = [img for img in os.listdir(class_path) 
                    if img.lower().endswith(('.png', '.jpg', '.jpeg'))]
        random.shuffle(images)

        split_point = int(len(images) * split)

        train_imgs = images[:split_point]
        test_imgs = images[split_point:]

        # Create class folders inside train/test
        os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
        os.makedirs(os.path.join(test_dir, class_name), exist_ok=True)

        # Copy to train
        for img in train_imgs:
            shutil.copy(
                os.path.join(class_path, img),
                os.path.join(train_dir, class_name, img)
            )

        # Copy to test
        for img in test_imgs:
            shutil.copy(
                os.path.join(class_path, img),
                os.path.join(test_dir, class_name, img)
            )

print("✅ Done! Train and Test folders created.")
```
